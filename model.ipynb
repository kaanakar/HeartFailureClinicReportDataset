```python
#from __future__ import absolute_import, division, print_function, unicode_literals

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc

import tensorflow as tf
from tensorflow import keras
```

```python
records = pd.read_csv("input\heart_failure_clinical_records_dataset.csv")
```

```python
records.head()
```

```python
records.shape
```

```python
train_records = records.iloc[:269,:]
eval_records = records.iloc[269:,:]
```

```python
train_records.shape
```

```python
y_train = train_records.pop('DEATH_EVENT')
y_eval = eval_records.pop('DEATH_EVENT')
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(8, activation='softmax'),
    keras.layers.Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_records, y_train, epochs=300)
```

```python
model.summary()
```

```python
test_loss, test_acc = model.evaluate(eval_records,  y_eval, verbose=1) 

print('Test accuracy:', test_acc)
```
